```python
#Data Cleaning
df = df.drop(['Id'],axis = 1)
```

**Label Encoding**

used to convert categorical data variables to numerical form, it helps machine learning algorithm to undertand handling the data correctly

```python
#x,y split
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 34)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#KNN model building
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=7,p=2 )
knn.fit(x_train, y_train)
```

```python
y_predict = knn.predict(x_test)
```

```python
y_predict
```

```python
print(len(y_test), len(y_predict))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_predict)
print('Confusion Matrix',confusion_matrix)
```

```python
from sklearn.metrics import accuracy_score


accuracy = accuracy_score(y_test,y_predict)
print('Accuracy',accuracy)
```

```python
#Building the AdaBoost Model Let's create the AdaBoost Model using Scikit-learn. AdaBoost uses Decision Tree Classifier as default Classifier.
```

```python
x1 = x.copy()
```

```python
x1_train, x1_test, y_train, y_test = train_test_split(x1, y, test_size=0.3,random_state = 25)
```

```python
# Create adaboost classifer object
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
# Train Adaboost Classifer
model = abc.fit(x1_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(x1_test)
```

```python
# Model Accuracy,
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```

```python
x2 = x.copy()
```

```python
x2_train, x2_test, y_train, y_test = train_test_split(x2, y, test_size=0.3,random_state = 25)
```

```python
# Load libraries
from sklearn.ensemble import AdaBoostClassifier

# Import Support Vector Classifier
from sklearn.svm import SVC
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
svc=SVC(probability=True, kernel='linear')

# Create adaboost classifer object
abc =AdaBoostClassifier(n_estimators=50, base_estimator=svc,learning_rate=1)

# Train Adaboost Classifer
model = abc.fit(x2_train, y_train)

#Predict the response for test dataset
y1_pred = model.predict(x2_test)


# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y1_pred))
```
